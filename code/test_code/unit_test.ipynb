```python
# 建立一個DataFrame
df = pd.DataFrame({
        'key': ['A', 'B', 'C', 'A', 'B', 'C'],
        'date':['2020-01-01', '2020-01-01','2020-01-01','2020-01-02','2020-01-02','2020-01-02'],
        'value': [1, 2, None, 4, None, 6]
    })

# 以 'key' 欄位為基礎，做前值填補
df['value'] = df.groupby(['key'])['value'].fillna(method='ffill')

print(df.reset_index(drop=True))
```

```python
# 生成一個範例資料集
df = pd.DataFrame({
    'group': ['A']*3 + ['B']*3,
    'value': [1, np.nan, 3, np.nan, 4, np.nan]
})

# 按照group排序
df = df.sort_values(by='group')

# 利用groupby和apply來進行前值填補
df['value_filled'] = df.groupby('group')['value'].apply(lambda x: x.fillna(method='ffill'))

print(df)
```

## 測試新流程:
- 把 tickers 迴圈移除: tickers 直接傳入 tejapi.get.coid
- 將 triggers-> transfer_language_columns模組化，面對長區間資料以一年為單位，分段執行
- by ticker 進行填補的方式改為 groupby tickers sort by date apply x.fillna(method='ffill')

```python
tickers=['2330', '2881', '2882']

data_org = tejapi.get('TWN/AFESTM1',
                    coid=tickers,
                    paginate=True,
                    chinese_column_name=False,
                    mdate={'gte': para.default_start, 'lte': para.default_end},
                    opts={'sort':{'coid.desc', 'mdate.desc'}})
data_org
```

```python
# select certain fin_type
data_sets = para.dask_mpf.get_certain_fin_type(data_org, ['A','Q','TTM'])
data_sets
```

```python
data_sets = para.dask_mpf.fin_pivot(data_sets, remain_keys=['coid', 'mdate', 'no', 'sem', 'annd', 'fin_type', 'curr', 'fin_ind'])
data_sets
```

```python
data_sets.query('mdate > annd')
```

```python
# transfer fin_ann_date to daily basis
days = para.dask_mpf.generate_multicalendars(tickers)
data_sets = dd.merge(days, data_sets, left_on=['coid','all_dates'], right_on=['coid', 'annd'], how='left')
data_sets
```

```python
data_sets = data_sets.sort_values(['coid', 'all_dates'])#.query('annd > all_dates')
```

```python
data_sets = data_sets.groupby('coid', group_keys=False).apply(para.dask_mpf.fillna_multicolumns, meta = data_sets)
```

```python
data_sets.compute().query('annd > all_dates')
```

```python
data_sets = data_sets.compute()
data_sets
```
