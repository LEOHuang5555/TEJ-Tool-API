```python
unique_date = get_announce_date(ticker=listOfStocks)
unique_date
```

```python
unique_date[unique_date['a_dd']=='2007-04-30']
```

```python
fin_data = pd.DataFrame()
for i in listOfStocks:
   fin = tejapi.get('TWN/AINVFINQ',
   coid = i,
   paginate = True,
   chinese_column_name=False,
      opts= {'pivot':True, 'columns':[]})
   fin_data = pd.concat([fin_data, fin])
fin_data
```

```python
# merge announce_date with financial report
fin_ann_date = fin_data.merge(unique_date, left_on=['coid', 'mdate', 'fin_od'], right_on= ['coid', 'mdate', 'fin_od'], how='left')
fin_ann_date
```

```python
# 營業日
days = pd.date_range(start='2005-01-01', end=datetime.datetime.now(), freq='D')
days = pd.DataFrame({'all_dates':days})
# 結合發布日(announce_date)以及營業日(bussiness days)
fin_BD = days.merge(fin_ann_date, left_on='all_dates', right_on = 'a_dd', how = 'left').ffill()
fin_BD # [df['年月']>=df['日曆日']] # [~df['公司碼'].isna()]
```

```python
x = fin_BD.groupby(['all_dates', 'coid', 'mdate']).count()['fin_od']
x[x > 1]
#fin_BD.loc[index]
```

```python
fin_BD[fin_BD['all_dates'] == '2007-04-30']
```

```python
merge_data = stk_price.merge(fin_BD, left_on=['coid', 'mdate'], right_on=['coid', 'all_dates'], how = 'left')#.ffill()#.fillna(method = 'backfill')
merge_data
```
